## Medidas de atribución de desempeño

En este apartado se comparan la inversión con la estrategia activa y la pasiva, ambas durante el periodo de pandemia.

Para poder comparar ambas estrategias se tienen 3 medidas:

- Rendimiento mensual promedio.
- Rendimiento mensual acumulado en el último periodo.
- Sharpe ratio.

**Sharpe ratio**

$$SharpeRatio = \frac{R_p - Rf}{\sigma_p}$$

Donde:

- $R_p$ Rendimiento del portafolio.
- $R_f$ Tasa libre de riesgo.
- $\sigma_p$ Desviación estándar del rendimiento del portafolio.


Para el Sharpe ratio se tomó la tasa libre de riesgo para México, anual al 01 de enero 2021 (0.0429).

De igual manera, para dicha medida se tomó el rendimiento mensual acumulado del último periodo para $R_p$


```python
Metrics().get_metrics()
```

**Nota:** Los rendimientos de la inversión activa pueden llegar a variar un poco respecto a los rendimientos que se muestran en el apartado de *Estrategia Activa en periodo de pandemia*. Esto es debido a que para el portafolio inicial al usar ponderaciones aleatorias para sacar el máximo sharpe ratio, puede llegar a variar un poco los resultados.

Al comparar ambas estrategias se puede observar que en las 3 medidas fue mejor la estrategia activa, debido a que tuvo mejores rendimiento y tuvo un mayor índice de Sharpe.

Esto es debido a que al rebalancear constantemente el portafolio, podemos aprovechar las subidas y bajadas de los precios a lo largo del tiempo, esto nos permite incrementar las ganancias o el capital del portafolio, aún teniendo los costos por comisiones.
